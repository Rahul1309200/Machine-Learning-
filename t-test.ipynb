```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
## 1. One-Sample T-Test
# A company claims the average delivery time is 30 minutes. We collected delivery times from 20 orders to verify.

delivery_times = pd.DataFrame({
    'order_id': range(1, 21),
    'delivery_time_min': [32, 28, 35, 30, 33, 29, 36, 31, 34, 27,
                          33, 35, 30, 32, 37, 28, 31, 34, 36, 33]
})
delivery_times
```

```python
claimed_mean = 31
t_stat, p_value = stats.ttest_1samp(delivery_times['delivery_time_min'], claimed_mean)
print(f"Sample Mean: {delivery_times['delivery_time_min'].mean():.2f}")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")

if(p_value < 0.05):
    print("Reject H0: There is enough difference between the sample mean and the claimed mean of 30 minutes.")
else:
    print("Fail to reject H0: There is not enough difference between the sample mean and the claimed mean of 30 minutes.")
```

```python
## 2. Independent Two-Sample T-Test
# Comparing exam scores of students from two different teaching methods.
exam_scores = pd.DataFrame({
    'method_A': [78, 85, 88, 72, 90, 76, 95, 80, 82, 87,
                 74, 91, 83, 79, 86],
    'method_B': [65, 70, 68, 72, 60, 75, 66, 71, 63, 69,
                 74, 67, 62, 73, 144]
})
exam_scores
```

```python
t_stat, p_value = stats.ttest_ind(exam_scores['method_A'], exam_scores['method_B'])
print(f"Mean A: {exam_scores['method_A'].mean():.2f}, Mean B: {exam_scores['method_B'].mean():.2f}")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value}")
if (p_value < 0.05):
    print("Reject H0: There is a significant difference in exam scores between the two teaching methods.")
else:
    print("Fail to reject H0: There is no significant difference in exam scores between the two teaching methods.")
```

```python
bp_data = pd.DataFrame({
    'patient_id': range(1, 16),
    'bp_before': [142, 138, 150, 145, 135, 148, 140, 155, 143, 137,
                  146, 152, 139, 141, 149],
    'bp_after':  [130, 132, 140, 138, 128, 135, 133, 145, 136, 130,
                  138, 142, 131, 134, 140]
})
bp_data
```

```python
t_stat, p_value = stats.ttest_rel(bp_data['bp_before'], bp_data['bp_after'])
print(f"Mean Before: {bp_data['bp_before'].mean():.2f}, Mean After: {bp_data['bp_after'].mean():.2f}")
print(f"T-statistic: {t_stat:.4f}, P-value: {p_value}")
if (p_value < 0.05):
    print("Reject H0: The medication has a significant effect on reducing blood pressure.")
else:    
    print("Fail to reject H0: There is no significant effect of the medication on blood pressure.")    
```

```python
## 4. Covariance
# Measures how two variables change together. Hours studied vs marks scored.
study_data = pd.DataFrame({
    'hours_studied': [2, 3, 5, 6, 7, 8, 9, 10, 11, 12],
    'marks_scored':  [45, 50, 60, 65, 70, 78, 82, 88, 92, 95]
})
study_data
```

```python
cov_matrix = study_data.cov()
print("Covariance Matrix:")
print(cov_matrix)
print(f"\nCovariance between hours & marks: {cov_matrix.loc['hours_studied', 'marks_scored']:.2f}")
print("Positive covariance → As hours increase, marks also increase.")
```

```python
## 5. Correlation (Pearson)
# Measures strength & direction of linear relationship. Height vs weight of 10 people.
body_data = pd.DataFrame({
    'height_cm': [150, 155, 160, 165, 170, 175, 180, 185, 190, 195],
    'weight_kg': [50, 55, 58, 62, 68, 72, 78, 82, 88, 95]
})
body_data
```

```python
corr, p_value = stats.pearsonr(body_data['height_cm'], body_data['weight_kg'])
print(f"Pearson Correlation: {corr:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Strong positive correlation" if corr > 0.7 else f"Weak/moderate correlation")
```

```python
## 6. Chi-Square Test of Independence
# Tests if two categorical variables are related. Gender vs preference for a product.
# Contingency table: rows = Gender (Male/Female), cols = Preference (Product A / Product B / Product C)
observed = pd.DataFrame({
    'Product_A': [30, 45],
    'Product_B': [50, 35],
    'Product_C': [20, 20]
}, index=['Male', 'Female'])
print("Observed Frequencies:")
observed
```

```python
chi2, p_value, dof, expected = stats.chi2_contingency(observed)
print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_value:.4f}")
print(f"\nExpected Frequencies:")
print(pd.DataFrame(expected, index=['Male', 'Female'], columns=['Product_A', 'Product_B', 'Product_C']))
print()
if p_value < 0.05:
    print("Reject H0: Gender and product preference are related.")
else:
    print("Fail to reject H0: No significant relationship between gender and product preference.")
```

```python
## 7. One-Way ANOVA
# Tests if means of 3 or more groups are significantly different. Crop yield with 3 fertilizers.
crop_data = pd.DataFrame({
    'fertilizer_A': [20, 22, 19, 24, 25, 21, 23, 18, 20, 22],
    'fertilizer_B': [28, 30, 27, 32, 29, 31, 26, 33, 28, 30],
    'fertilizer_C': [35, 38, 40, 36, 42, 39, 37, 41, 34, 38]
})
crop_data
```

```python
f_stat, p_value = stats.f_oneway(crop_data['fertilizer_A'], crop_data['fertilizer_B'], crop_data['fertilizer_C'])
print(f"Mean A: {crop_data['fertilizer_A'].mean():.2f}, Mean B: {crop_data['fertilizer_B'].mean():.2f}, Mean C: {crop_data['fertilizer_C'].mean():.2f}")
print(f"F-statistic: {f_stat:.4f}, P-value: {p_value}")
if p_value < 0.05:
    print("Reject H0: At least one fertilizer has a significantly different yield.")
else:
    print("Fail to reject H0: No significant difference in yield among the three fertilizers.")
```
